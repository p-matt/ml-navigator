```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/dataset_01.csv")
```

```python
df = df.sample(frac=1)
df.head()
```

```python
split_ratio = int(df.shape[0] * .95)
split_ratio
```

```python
df_train = df.iloc[:split_ratio, :]
df_inferences = df.iloc[split_ratio:, :]

df_train.shape, df_inferences.shape
```

```python
df_train.to_csv("dataset_01_train")
df_inferences[df_inferences.columns[~df_inferences.columns.isin(["price"])]].to_csv("dataset_01_inferences_X")
df_inferences[["price", "car_ID"]].to_csv("dataset_01_inferences_y")
```
